```
estudiante_1 = 58   # Nota baja (Cerca del límite)
estudiante_2 = 72   # Nota media-baja
estudiante_3 = 85   # Nota media-alta
estudiante_4 = 92   # Nota sobresaliente
estudiante_5 = 100  # Nota perfecta
puntajes_examen = [estudiante_1, estudiante_2, estudiante_3, estudiante_4, estudiante_5]
def obtener_calificacion(nota):
    if nota >=90:
        return("A")
    elif nota>=80:
        return("B")
    elif nota>=70:
        return("C")
    elif nota>=60:
        return("D")
    else:
        return("F")
for i, p in enumerate (puntajes_examen, start=1):
  calificacion = obtener_calificacion(p)
  print(f"Estudiante {i}: Puntaje {p} - Calificación {calificacion}")
```

```
# Definimos las notas con nombres propios
esteban = 58
maria = 72
juan = 85
ana = 92
luis = 100
puntajes_examen = [esteban, maria, juan, ana, luis]
nombres_estudiantes = ["Esteban", "María", "Juan", "Ana", "Luis"]
def obtener_calificacion(nota):
    if nota >=90:
        return("A")
    elif nota>=80:
        return("B")
    elif nota>=70:
        return("C")
    elif nota>=60:
        return("D")
    else:
        return("F")
for nombre_alumno, nota_num in zip (nombres_estudiantes, puntajes_examen):
  calificacion = obtener_calificacion(nota_num)
  print(f"Estudiante {nombre_alumno}: Puntaje {nota_num} - Calificación {calificacion}")
```
