```
#Para imprimir con dos numeros decimales
import random
precio = random.uniform(50,100)
print(precio)
print(f"El precios es $ {precio:.2f}")

#para un solo decimal uso :.1f
porcentaje = random.uniform(70,90)
print(porcentaje)
print(f"El porcentaje completado es {porcentaje:.1f}%")

#Para formatear fechas
from datetime import datetime
fecha_actual = datetime.now()
print(fecha_actual)
print(f"la fecha de hoy es {fecha_actual:%d/%m/%Y %H:%M}")
```
