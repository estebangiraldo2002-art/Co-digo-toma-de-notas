```
# Bucles con While: Bucles que solo se detienen cuando una condición no se cumple

import random
inventario = {"Teclado": 25, "Mouse": 18, "Monitor": 150, "Cargador": 12, "Audífonos": 35}

def consultar_precio(producto_buscado):
  if producto_buscado in inventario:
    return f"El {producto_buscado} cuesta {inventario[producto_buscado]}"
  else:
    return f"Lo sentimos, no vendemos {producto_buscado}" #En toda la parte anterior creé la función

continuar=True #Declaro una variable para usar el While (un bucle que se detiene hasta que la condición no se cumpla)

while continuar: #Aquí le digo que mientras continuar sea True entonces:
  producto=input("Ingrese el nombre del producto que desea consultar o 'salir' para terminar: ") #le pido al usuario que me escriba lo que está buscando
                                                                                                 #y lo tomo como input para mi variable
  if producto=="salir":
    continuar=False
  else:
    resultado=consultar_precio(producto)
    print(resultado)
```

```
# Bucles con While: Bucles que solo se detienen cuando una condición no se cumple - Consola estilizada

import random
inventario = {"Teclado": 25, "Mouse": 18, "Monitor": 150, "Cargador": 12, "Audífonos": 35}

def consultar_precio(producto_buscado):
  if producto_buscado in inventario:
    return f"El {producto_buscado} cuesta {inventario[producto_buscado]}"
  else:
    return f"Lo sentimos, no vendemos {producto_buscado}" #En toda la parte anterior creé la función

continuar=True #Declaro una variable para usar el While (un bucle que se detiene hasta que la condición no se cumpla)

while continuar: #Aquí le digo que mientras continuar sea True entonces:
  producto=input("Ingrese el nombre del producto que desea consultar o 'salir' para terminar: ").strip().capitalize() #le pido al usuario que me escriba lo que está buscando
                                                                                                 #y lo tomo como input para mi variable
  if producto=="Salir":
    print("-"*30)
    print("👋 ¡Gracias por usar el sistema!")
    print("-" * 30)
    continuar=False
  else:
    print("-" * 30)
    resultado = consultar_precio(producto)
    print(resultado)
    print("-" * 30)
```

# Proyecto Carrito de compras - Gemini

```
#Carrito con descuentos y ticket

inventario = {"Teclado": 25, "Mouse": 18, "Monitor": 150, "Cargador": 12, "Audífonos": 35}
inventario_precios = {"Teclado": 35.99, "Mouse": 15.50, "Monitor": 199.99, "Cargador": 20.00, "Audífonos": 55.00}
carrito=[]
total_bruto=0
iva=0.19
envio=1.99
monto_minimo_descuento= 80

descuento_si_vip=0.05
descuento_si_monto=0.05

continuar=True

while continuar:
  producto=input("Ingrese el nombre del producto que desea agregar al carrito o 'Terminar' para confirmar su carrito: ").strip().capitalize()
  if producto=="Terminar":
    if len(carrito)>0:
      print("-"*30)
      print("Carrito terminado 🛒")
      print("-" * 30)
    continuar=False
  else:
    if producto in inventario:
      carrito.append(producto)
      print(f"✅ {producto} agregado al carrito")
      total_bruto += inventario_precios[producto]
    else:
      print(f"Lo sentimos, no vendemos {producto}")

# --- AQUÍ EMPIEZA EL BLOQUE QUE ORGANIZA TODO ---
if len(carrito) > 0:
    # 1. Pregunta VIP (Solo si compró algo)
    continuar_vip = True
    while continuar_vip:
        vip = input("¿Es cliente VIP? (Si/No): ").strip().capitalize()
        if vip == "Si":
            print("-" * 30)
            print("✅ Cliente VIP")
            print("-" * 30)
            continuar_vip = False
        elif vip == "No":
            print("-" * 30)
            print("❌ No es cliente VIP")
            print("-" * 30)
            continuar_vip = False
        else:
            print("❌ Opción no válida")

    # 2. Cálculos de descuentos (Movidos adentro para evitar errores)
    if vip == "Si":
        descuento_vip = descuento_si_vip
    else:
        descuento_vip = 0

    if total_bruto >= monto_minimo_descuento:
        descuento_monto = descuento_si_monto
    else:
        descuento_monto = 0

    monto_descuento_vip = descuento_vip * total_bruto
    monto_descuento_monto = descuento_monto * total_bruto
    descuento_total = descuento_vip + descuento_monto
    total_descuento = total_bruto * descuento_total
    valor_iva=(total_bruto - total_descuento) * (iva)
    subtotal=(total_bruto-total_descuento)
    total_pagar = ((total_bruto - total_descuento) + valor_iva) + envio

    # 3. Tu Ticket Formateado (Tal como lo tenías, pero con la sangría correcta)
    print("-" * 30)
    print("Resumen de compra:")
    print("-" * 30)
    print(f"\nProductos en el carrito: \n")
    for producto in carrito:
        precio_unitario = inventario_precios[producto]
        print(f"{producto}: ${precio_unitario:.2f}")

    print(f"Subtotal productos: {total_bruto:.2f}\n")

    if descuento_monto > 0 or descuento_vip > 0:
        print("-" * 30)
        print("\n✨ DESCUENTOS APLICADOS:\n")
        if descuento_vip > 0:
            # Corregido aquí para que use la variable de porcentaje VIP
            print(f"Cliente VIP {descuento_si_vip*100:.0f}%: ${monto_descuento_vip:.2f}")
        if descuento_monto > 0:
            print(f"Descuento por monto de compra {descuento_si_monto*100:.0f}%: ${monto_descuento_monto:.2f}")
        print(f"\n","-" * 30)
    else:
        print("-" * 30)
        print("\nNo hay descuentos aplicables\n")
        print("-" * 30)

    print(f"Subtotal: ${subtotal:.2f}")
    print(f"Valor IVA: ${valor_iva:.2f}")

    print(f"\nTotal a pagar: ${total_pagar:.2f}")
    print(f"\n","-" * 30)

else:
    # 4. Si el carrito está vacío, termina de inmediato aquí
    print("-" * 30)
    print("El carrito está vacío. ¡Vuelve pronto!")
    print("-" * 30)
```
